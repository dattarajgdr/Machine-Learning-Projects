### Importing the required python libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading comanies.csv into companies and rounds2.csv into rounds2 Dataframe

```python
companies=pd.read_csv(r'C:\Users\Dell\Documents\IIIT-B\Inferential Statistics\companies.csv',encoding='palmos')
rounds2=pd.read_csv(r'C:\Users\Dell\Documents\IIIT-B\Inferential Statistics\rounds2.csv',encoding='palmos')
```

Converting the Unique column values to uppercase

```python
companies.permalink=companies.permalink.str.upper()
rounds2.company_permalink=rounds2.company_permalink.str.upper()
```

### To find unique companies present in the companies file

### To check whether all the round2 companies are present in companies dataset

### Merging both the dataframes

```python
master_frame=pd.merge(companies,rounds2,left_on='permalink',right_on='company_permalink')
master_frame.shape
```

```python
master_frame.head()
```

### Dropping permalink from merged data set as it is a duplicate column

```python
master_frame=master_frame.drop('permalink',axis=1)
master_frame
```

### Lets check for missing values for data cleaning

```python
master_frame.isnull().sum()
```

#### As there are many nulls in the homepage_url,state_code,region,city and founded_at,funding_round_code we can drop these columns as part of data cleaning. But we cannot replace or induce values in category_list,funding_round_code,raised_amount_usd because we need to derive based on these columns

```python
master_frame=master_frame.drop(['homepage_url','state_code','region','city','founded_at','funding_round_code'],axis=1)
```

```python
master_frame.isnull().sum()
```

As there are many records with nulls in the raised_amount_usd we cannot impute them. So it is better to drop those rows else might impact our analysis.

```python
master_frame=master_frame[~np.isnan(master_frame.raised_amount_usd)]
```

```python
master_frame.isnull().sum()
```

As we can see there are still 3 columns having many null rows to be handles. Now lets drop nulls in country_code.

```python
master_frame=master_frame[~pd.isnull(master_frame.country_code)]
```

```python
master_frame.isnull().sum()
```

Now lets handle the null values in category_list by removing those rows

```python
master_frame=master_frame[~pd.isnull(master_frame.category_list)]
master_frame.isnull().sum()
```

As there is only 1 null record in name there is no need to handle it because its negligible.

Loading Mapping Data

```python
mapping=pd.read_csv(r'C:\Users\Dell\Documents\IIIT-B\Inferential Statistics\mapping.csv',encoding='palmos')
mapping.head()
```

```python
mapping=mapping[~pd.isnull(mapping['category_list'])]
mapping['category_list']=mapping['category_list'].str.lower()
mapping.head()
```

We can see that there are 0's between the category list names. This looks like it might be happened when handling na values,So lets replace it with na.

```python
mapping[~mapping['category_list'].isin(venture_funding['primary sector'])]
```

```python
mapping['category_list']=mapping['category_list'].apply(lambda x: x.replace('0','na'))
```

```python
mapping_melted=pd.melt(mapping, id_vars=['category_list'],value_vars=['Automotive & Sports',
       'Cleantech / Semiconductors', 'Entertainment', 'Health',
       'Manufacturing', 'News, Search and Messaging', 'Others',
       'Social, Finance, Analytics, Advertising'])
```

```python
mapping_melted=mapping_melted[mapping_melted.value==1]
```

```python
mapping_melted=mapping_melted.drop(columns='value')
mapping_melted.shape
```

```python
mapping_melted=mapping_melted.rename(columns={'category_list':'primary sector'})
```

```python
mapping_melted.head(30)
```

```python
merge_venture_mapping=pd.merge(venture_funding,mapping_melted,on='primary sector',how='inner')
```

### Plots

Pie Plot to display the fraction of total investments in venture,seed,angel and private_equity

```python
plot_master_frame=master_frame[master_frame.funding_round_type.isin(['venture', 'seed', 'angel', 'private_equity'])]
plt.title('Fraction of Total Investments')
plot_master_frame.funding_round_type.value_counts().plot.pie(autopct='%1.2f%%')
plt.show()
```

Bar plot to display the average amount of investment in Millions USD

```python
sns.set(style="whitegrid")
sns.barplot(x=plot_master_frame['funding_round_type'],y=plot_master_frame['raised_amount_usd']/1000000,estimator=np.mean,ci=None)
plt.title("Average Amount of Investement") 
plt.ylabel("raised_amount_usd(in millions)")
plt.show()
```

Top 9 countries with total amount of investment in venture

```python
top9['English_speaking']=top9.country_code.isin(['USA','GBR','IND','CAN','ISR'])
sns.barplot(x=top9.country_code,y=top9.raised_amount_usd/1000000,hue=top9.English_speaking,ci=None,dodge=False)
plt.show()
```
